```python
def createdataset(dataset,start,stop):
    count=0
    for k in range(start,stop):
        c=(20*k)
        for i in range(c,c+16):
            for j in range(16):    
                y=(j-1)%16
                if i==c:
                    trainset.loc[count]=[dataset.ix[i-25,y%16],dataset.ix[i-25,(y+1)%16],dataset.ix[i-25,(y+2)%16],dataset.ix[i-40,y%16],dataset.ix[i-40,(y+1)%16],dataset.ix[i-40,(y+2)%16],dataset.ix[i-39,y%16],dataset.ix[i-39,(y+1)%16],dataset.ix[i-39,(y+2)%16],dataset.ix[i,j]]     
                elif i==(c+15):
                    trainset.loc[count]=[dataset.ix[i-41,y%16],dataset.ix[i-41,(y+1)%16],dataset.ix[i-41,(y+2)%16],dataset.ix[i-40,y%16],dataset.ix[i-40,(y+1)%16],dataset.ix[i-40,(y+2)%16],dataset.ix[i-55,y%16],dataset.ix[i-55,(y+1)%16],dataset.ix[i-55,(y+2)%16],dataset.ix[i,j]] 
                else:  
                    trainset.loc[count]=[dataset.ix[i-41,y%16],dataset.ix[i-41,(y+1)%16],dataset.ix[i-41,(y+2)%16],dataset.ix[i-40,y%16],dataset.ix[i-40,(y+1)%16],dataset.ix[i-40,(y+2)%16],dataset.ix[i-39,y%16],dataset.ix[i-39,(y+1)%16],dataset.ix[i-39,(y+2)%16],dataset.ix[i,j]] 
                count=count+1
                
    return trainset
trainset=createdataset(dataset,2,52)
        
```
